# Begin Problem 2 - School Summary

```python
# Combine the data into a single dataset.  This will allow combining the grades with the schools.
schoolWithScores = pd.merge(school_data, student_data, how="left", on=["school_name", "school_name"])
schoolWithScores.head(30)
```

```python
#delete the columns not needed
del schoolWithScores['School ID']
del schoolWithScores["Student ID"]
del schoolWithScores["student_name"]
del schoolWithScores["gender"]
del schoolWithScores["grade"]
```

```python
schoolWithScores.head(3)
```

```python
#Add columns to indicate the number of students who are passing math and reading

schoolWithScores["Passing Reading?"] = np.where(schoolWithScores["reading_score"] >= 70, 1, 0)
schoolWithScores["Passing Math?"] = np.where(schoolWithScores["math_score"] >= 70, 1, 0)
schoolWithScores.head(5)
```

```python
#Save schoolWithScores to a new df where I will define how each column is aggregated into the new df
dfSchoolGroupAgg = schoolWithScores.groupby("school_name").agg({"size":'mean',
                                                                 "budget":'mean',
                                                                 "math_score":'mean',
                                                                 "reading_score":'mean',
                                                                 "Passing Math?" : 'sum',
                                                                 "Passing Reading?": 'sum'})

dfSchoolGroupAgg.head(20)
```

```python
#From the aggregated list, add the needed columns
dfSchoolGroupAggPlus = dfSchoolGroupAgg
```

```python
#Add the missing columns

dfSchoolGroupAggPlus["Per Student Budget"] = dfSchoolGroupAggPlus["budget"] / dfSchoolGroupAggPlus["size"]
dfSchoolGroupAggPlus["%Percent Passing Reading"] = dfSchoolGroupAggPlus["Passing Reading?"] / dfSchoolGroupAggPlus["size"]
dfSchoolGroupAggPlus["%Percent Passing Math"] = dfSchoolGroupAggPlus["Passing Math?"] / dfSchoolGroupAggPlus["size"]
dfSchoolGroupAggPlus["Overall Pass Rate"] = (dfSchoolGroupAggPlus["%Percent Passing Reading"] + dfSchoolGroupAggPlus["%Percent Passing Math"]) / 2

dfSchoolGroupAggPlus.head(20)
```

```python
dfSchoolGroupUnclean = dfSchoolGroupAggPlus
dfSchoolGroupUnclean.head(20)
```

```python
dfSchoolGroupUnclean.columns
#need a list of column names to ensure the correct ones are removed
```

```python
#delete the columns not needed
del dfSchoolGroupUnclean['Passing Math?']
del dfSchoolGroupUnclean["Passing Reading?"]

#rename the column headers
dfSchoolGroupUnclean = dfSchoolGroupUnclean.rename(columns={"size": "Total Students", 
                                              "budget": "Total Budget",
                                              "math_score": "Average Math Score",
                                              "reading_score": "Average Reading Score"
                                             })
dfSchoolGroupUnclean.head(20)
```

```python
#The only missing column is the school type.  Create a merge to get this data from the original school_date

dfSchoolDataType = pd.merge(dfSchoolGroupUnclean, school_data, how="left", on=["school_name", "school_name"])
```

```python
dfSchoolDataType.head()
```

```python
#now keep only those columns needed and set them in the order required by the new dataframe

dfSchoolDataOrdered = dfSchoolDataType[["school_name", "type", "Total Students", "Total Budget", "Per Student Budget", "Average Math Score",
                                       "Average Reading Score", "%Percent Passing Math", "%Percent Passing Reading", "Overall Pass Rate"]]
```

```python
dfSchoolDataOrdered.head(20)
```

```python
#Rename the column headers from the merge
dfSchoolDataOrdered = dfSchoolDataOrdered.rename(columns={"school_name": "School Name", 
                                              "type": "Type"
                                             }) 

dfSchoolDataOrdered.head(20)
```

```python
dfSchoolDataFormattedFinal = dfSchoolDataOrdered   #keep a copy of the unformatted values...just in case I need to look back at these or use them in an upcoming example.
```

```python
#Add user friendly formatting
dfSchoolDataFormattedFinal["Total Students"] = dfSchoolDataFormattedFinal["Total Students"].astype(float).map("{:,.0f}".format)
dfSchoolDataFormattedFinal["Total Budget"] = dfSchoolDataFormattedFinal["Total Budget"].astype(float).map("${:,.2f}".format)
dfSchoolDataFormattedFinal["Per Student Budget"] = dfSchoolDataFormattedFinal["Per Student Budget"].astype(float).map("${:,.2f}".format)
dfSchoolDataFormattedFinal["Average Math Score"] = dfSchoolDataFormattedFinal["Average Math Score"].astype(float).map("{:,.4f}".format)
dfSchoolDataFormattedFinal["Average Reading Score"] = dfSchoolDataFormattedFinal["Average Reading Score"].astype(float).map("{:,.4f}".format)

dfSchoolDataFormattedFinal["%Percent Passing Math"] = dfSchoolDataFormattedFinal["%Percent Passing Math"].astype(float).map("{:,.3%}".format)
dfSchoolDataFormattedFinal["%Percent Passing Reading"] = dfSchoolDataFormattedFinal["%Percent Passing Reading"].astype(float).map("{:,.3%}".format)
dfSchoolDataFormattedFinal["Overall Pass Rate"] = dfSchoolDataFormattedFinal["Overall Pass Rate"].astype(float).map("{:,.3%}".format)

dfSchoolDataFormattedFinal.head(20)
```
